```
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread('j_ruido_branco.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
plt.imshow(erosion, 'gray')
plt.title('Erosão')
plt.show()

dilate = cv2.dilate(erosion,kernel,iterations = 1)
plt.imshow(dilate, 'gray')
plt.title('Dilatação')
plt.show()
```

```
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread('j_ruido_branco.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, 'gray')
plt.title('Abertura =  erosão + dilatação')
plt.show()
```

```
from matplotlib import pyplot as plt
import cv2
import numpy as np

img = cv2.imread('j_ruido_escuro.png',0)
kernel = np.ones((11,11),np.uint8)
closing  = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
plt.imshow(closing, 'gray')
plt.title('Fechamento = dilatação + erosão')
plt.show()
```

```
from matplotlib import pyplot as plt
import cv2
import numpy as np
%matplotlib inline

img = cv2.imread('Cenario.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(T, bin) = cv2.threshold(img, 140, 255, cv2.THRESH_BINARY)
plt.subplot(1,3,1)
plt.imshow(bin, 'gray')
plt.title('Binarizada')

kernel = np.ones((5,5),np.uint8)
dilate = cv2.dilate(bin,kernel,iterations = 1)
plt.subplot(1,3,2)
plt.imshow(dilate,'gray')
plt.title('Dilatação')


erode = cv2.erode(dilate,kernel,iterations = 1)
plt.subplot(1,3,3)
plt.imshow(erode,'gray')
plt.title('Erosão')

plt.show()
```
